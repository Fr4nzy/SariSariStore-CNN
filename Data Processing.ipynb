Process the data by converting them into jpg and resize it to 224, 224

```python
import os
import cv2
from PIL import Image
```

# Old PreProcessor

```python
# Set input and output directories
input_directory = r"C:\Users\mzowe\Documents\SSSDataSets\ImageCNN\ProductsDataSets"
output_directory = r"C:\Users\mzowe\Documents\SSSDataSets\ImageCNN\ProcessedProductsDataSets"

# Function to resize and convert images
def resize_and_convert(input_path, output_path, target_size=(224, 224)):
    image = cv2.imread(input_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    image = cv2.resize(image, target_size)  # Resize image to target size

    # Save image as JPG
    output_file = os.path.join(output_path, os.path.basename(input_path))
    output_file = os.path.splitext(output_file)[0] + ".jpg"
    Image.fromarray(image).save(output_file, "JPEG")

# Iterate through the input directory
for root, dirs, files in os.walk(input_directory):
    for file in files:
        if file.lower().endswith((".png", ".jpg")):
            input_path = os.path.join(root, file)
            output_subdirectory = os.path.join(output_directory, os.path.relpath(root, input_directory))
            
            # Create the output subdirectory if it doesn't exist
            if not os.path.exists(output_subdirectory):
                os.makedirs(output_subdirectory)
            
            # Resize and convert the image
            resize_and_convert(input_path, output_subdirectory)
            
            print(f"Processed: {input_path}")

print("Processing complete.")
```

Data Splitting

```python
import os
import shutil
import random

# Paths to the dataset and output directories
dataset_dir = r"C:\Users\mzowe\Documents\SSSDataSets\ImageCNN\ProcessedProductsDataSets"
output_dir = r"C:\Users\mzowe\Documents\SSSDataSets\ImageCNN\SplittedProductsDataSets"

train_ratio = 0.70  # 70% for training
val_ratio = 0.15    # 15% for validation
test_ratio = 0.15    # 15% for testing

# Create output directories
os.makedirs(os.path.join(output_dir, "train"), exist_ok=True)
os.makedirs(os.path.join(output_dir, "validation"), exist_ok=True)
os.makedirs(os.path.join(output_dir, "test"), exist_ok=True)

# Iterate through class folders in the dataset directory
for class_name in os.listdir(dataset_dir):
    class_dir = os.path.join(dataset_dir, class_name)

    # Get a list of all image files in the class directory
    image_files = [f for f in os.listdir(class_dir) if f.endswith((".jpg", ".jpeg"))]

    # Shuffle the image files randomly
    random.shuffle(image_files)

    # Calculate the split indices
    num_images = len(image_files)
    num_train = int(num_images * train_ratio)
    num_val = int(num_images * val_ratio)

    # Split the images into train, validation, and test sets
    train_images = image_files[:num_train]
    val_images = image_files[num_train:num_train + num_val]
    test_images = image_files[num_train + num_val:]

    # Process and copy the images to their respective directories
    for split, images in [("train", train_images), ("validation", val_images), ("test", test_images)]:
        for image in images:
            src_path = os.path.join(class_dir, image)
            dest_path = os.path.join(output_dir, split, class_name, image)
            os.makedirs(os.path.dirname(dest_path), exist_ok=True)
            shutil.copy(src_path, dest_path)

print("Splitting and copying completed.")
```
