```python
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import requests
import time
from PIL import Image
from io import BytesIO

# Define search parameters
google_search_query = "Del Monte Original Style Tomato Sauce"
google_limit = 150
google_output_directory = "C://Users//mzowe//Documents//SSSDataSets//ImageCNN//Products and Barcode//Other Products"

# Create a subfolder for the downloaded images
output_subfolder = os.path.join(google_output_directory, google_search_query)

# Initialize the Chrome WebDriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # Optional: Run headless without a browser window
chrome_options.add_argument('--disable-extensions')
chrome_options.add_argument('--disable-popup-blocking')

# Set the path to your Chrome executable (change it to your Chrome installation path)
chrome_options.binary_location = 'C:/Program Files/Google/Chrome/Application/chrome.exe'

# Create a Chrome WebDriver instance
browser = webdriver.Chrome(options=chrome_options)

# Function to collect image URLs with dimensions not lower than 100x100
def collect_image_urls(query, limit, min_width=100, min_height=100):
    search_url = f"https://www.google.com/search?q={query}&tbm=isch"

    # Navigate to the Google Images page
    browser.get(search_url)

    image_urls = set()  # Use a set to store unique URLs
    last_height = 0

    while len(image_urls) < limit:
        # Scroll down to the end of the page
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait for new images to load
        time.sleep(2)  # Adjust the sleep time if needed

        # Get the updated page source and parse it
        page_source = browser.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        # Find image URLs in the page source
        for img in soup.find_all('img', src=True):
            src = img['src']
            if src.startswith('http'):
                # Check image dimensions
                try:
                    response = requests.get(src)
                    if response.status_code == 200:
                        img_data = Image.open(BytesIO(response.content))
                        width, height = img_data.size
                        if width >= min_width and height >= min_height:
                            image_urls.add(src)
                except Exception as e:
                    pass

        # If no new images were loaded, exit the loop
        new_height = browser.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    return list(image_urls)[:limit]

# Collect image URLs with dimensions not lower than 100x100
image_urls = collect_image_urls(google_search_query, google_limit, min_width=100, min_height=100)

# Function to download images
def download_images(image_urls, output_directory, image_formats):
    os.makedirs(output_directory, exist_ok=True)  # Create the subfolder for downloaded images
    
    for i, image_url in enumerate(image_urls):
        try:
            response = requests.get(image_url)
            if response.status_code == 200:
                with open(os.path.join(output_directory, f"{google_search_query}_{i + 1}.{image_formats[0]}"), 'wb') as img_file:
                    img_file.write(response.content)
                print(f"Downloaded image {i + 1}")
            else:
                print(f"Error downloading image {i + 1}: HTTP Status {response.status_code}")
        except Exception as e:
            print(f"Error downloading image {i + 1}: {str(e)}")

# Download the collected images to the subfolder
download_images(image_urls, output_subfolder, google_formats)

# Close the Chrome WebDriver when done
browser.quit()
```
